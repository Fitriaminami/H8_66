# uniform distribution

```python
from scipy.stats import uniform
```

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size=n, loc = start, scale=width)
```

# Confidence Intervals

```python
import pandas as pd
import numpy as np
df = pd.read_csv('Heart.csv')
df
```

```python
df['Sex1'] = df.sex.replace({1:"Male", 0:"Female"})
```

```python
dx = df[["target", "Sex1"]].dropna()
```

```python
pd.crosstab(dx.target, dx.Sex1)
```

```python
p_fm = 226/(86+226)
p_fm
```

```python
n = 86+226
n # populasi
```

```python
se_female = np.sqrt(p_fm * (1-p_fm)/n)
se_female # standart error
```

```python
z_score = 1.96
lcb = p_fm - z_score* se_female
ucb = p_fm + z_score* se_female
lcb, ucb # confidence interval
```

```python
import statsmodels.api as sm
sm.stats.proportion_confint(n*p_fm, n)
```

# Hypothesis Testing

```python
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# One Population Proportion

- Null Hypotesis = 52%
- Alt Hypotesis => 52%

```python
n = 1018
pnull = .52
phat = .56
```

```python
sm.stats.proportions_ztest(phat * n, n, pnull, alternative='larger')
```

```python
# karena nilai p-vaue dari z-test cukup kecil, kita dapat reject the Null Hypothesis bahwa persentase orang tua yang percaya bahwa kurang tidur remaja mereka disebabkan oleh elektronik sama dengan tahun -tahun sebelumnya. yaitu sekitar 52%
```

# Difference in Population Means

```python
url = "https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/nhanes_2015_2016.csv"
da = pd.read_csv(url)
da.head()
```

```python
females = da[da["RIAGENDR"] == 2]
male = da[da["RIAGENDR"] == 1]
```

```python
n1 = len(females)
mu1 = females["BMXBMI"].mean()
sd1 = females["BMXBMI"].std()

(n1, mu1, sd1)
```

```python
n2 = len(male)
mu2 = male["BMXBMI"].mean()
sd2 = male["BMXBMI"].std()

(n2, mu2, sd2)
```

```python
sm.stats.ztest(females["BMXBMI"].dropna(),
male["BMXBMI"].dropna(),alternative='two-sided')    
```

```python
# karena hasil p-value (6,59*10^-10) sangat kecil, maka kita bisa reject null hypothesis (BMI laki=BMI perempuan)
```

```python
plt.figure(figsize=(7,4))
plt.title("Female BMI histogram", fontsize=16)
plt.hist(females["BMXBMI"].dropna(),edgecolor='k', color='pink', bins=25)
plt.show

plt.figure(figsize=(7,4))
plt.title("Male BMI histogram", fontsize=16)
plt.hist(male["BMXBMI"].dropna(),edgecolor='k', color='blue', bins=25)
plt.show
```

# Hypothesis Testing for One Proportion

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats.distributions as dist
df = pd.read_csv('Heart.csv')
df.head()
```

- H0: p0 = 0.42
- Ha: p > 0.42

```python
p_us = len(df[df['target']=='1'])/len(df)
```

```python
# standart error
se = np.sqrt(0.42 * (1-0.42)/len(df))
```

```python
be = p_us

he = 0.42
test_stat = (be-he)/se
test_stat
```

```python
pvalue = 2*dist.norm.cdf(-np.abs(test_stat))
pvalue
```
