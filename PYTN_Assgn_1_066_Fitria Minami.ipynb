# Area

```python
df5 = pd.pivot_table(df, values='value', index=['borough'],columns=['year'], aggfunc=np.sum)
```

```python
df5['Total']=df5.sum(axis=1) 
```

```python
df5
```

```python
df5.sort_values(by="Total", ascending=False, axis=0, inplace=True)
df_top5 = df5.head(5)
```

```python
df_top5.head()
```

```python
df_top5.drop("Total", axis=1, inplace=True)
```

```python
df_top5
```

```python
df_top5 = df_top5.transpose()
```

```python
df_top5
```

```python
df_top5.index = df_top5.index.map(int)
```

```python
df_top5.plot(kind="area",
            stacked=False,
            figsize=(20,10)
            )

plt.title("5 Wilayah dengan Kasus Kejahatan Tertinggi di London selama 2008-2016")
plt.ylabel("Jumlah Kasus")
plt.xlabel("Tahun")
```

# Pie Chart

```python
df6=df.groupby('borough', as_index=True).agg({"value":"sum"}).sort_values(by=['value'],ascending=[False])
```

```python
df6.plot.pie(y='value', 
                    figsize=(15,6),
                   autopct='%1.1f%%',
                   startangle=90,
                   shadow=True,
                   labels=None,
                   pctdistance=1.12)

plt.title("Sebaran Kasus Kejahatan di London Berdasarkan Wilayah selama tahun 2008-2016")
plt.axis('equal')

plt.show()
```

```python
df7=df.groupby('major_category', as_index=True).agg({"value":"sum"}).sort_values(by=['value'],ascending=[False])
```

```python
df7.plot.pie(y='value', 
                    figsize=(15,6),
                   autopct='%1.1f%%',
                   startangle=90,
                   shadow=True,
                   labels=None,
                   pctdistance=1.12)

plt.title("Jenis Kejahatan di London")
plt.axis('equal')

plt.show()
```
