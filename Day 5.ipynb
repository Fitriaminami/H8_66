```python
df.head()
```

```python
current_decade = df[df["year_id"] > 2010]
```

```python
current_decade
```

```python
games_with_notes = df[df["notes"].notnull()]
```

```python
games_with_notes.shape
```

```python
df[
    (df["_iscopy"] == 0)&
    (df["pts"] > 100)&
    (df["opp_pts"] > 100) &
    (df["team_id"] == "BLB")
]
```

```python
df[
    (df["_iscopy"] == 0)&
    (df["team_id"].str.startswith("LA")) &
    (df["year_id"] == 1992) &
    (df["notes"].notnull())
]
```

```python
df[df.duplicated(subset=['game_id', 'date_game'])]
```
